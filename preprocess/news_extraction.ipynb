```python
import numpy as np
import pandas as pd
import re
import requests
from datetime import datetime, timedelta
# from newsapi import NewsApiClient

api_key = ''
dtformat = '%Y-%m-%d'
# end_time = datetime(2022, 12, 31)
# start_time = datetime(2019, 12, 1)
end_time = datetime(2020, 3, 29)
start_time = datetime(2020, 3, 26)
keyword = 'finance OR Hong Kong'
sources = 'cbs-news,reuters,cnn,the-washington-post,the-washington-times,cbs-news,associated-press,bloomberg'
day = 1
df = pd.DataFrame()

# newsapi = NewsApiClient(api_key=api_key)

whitespace = re.compile(r"\s+")
num = re.compile(r"[0-9]")
mention = re.compile(r"@\S+")
hashtag = re.compile(r"#\s+")
word_in_bracket = re.compile(r"(\(.*\))|(\[.*\])")
web_address = re.compile(r"(?i)https(s):\/\/[a-z0-9.~_\-\/]+")
special_char = re.compile(r"[^a-zA-Z0-9\s]+")

# News Date Transformation
def go_back(now, day):
    return now - timedelta(days=day)

# Data Cleansing for News
def clean(news):
  news = whitespace.sub(' ', news)
  news = num.sub('', news)
  news = mention.sub('', news)
  news = hashtag.sub('', news)
  news = word_in_bracket.sub('', news)
  news = web_address.sub('', news)
  news = special_char.sub('', news)
  return news
```
