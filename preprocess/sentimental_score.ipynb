```python
import pandas as pd
import numpy as np
import os
import glob
import flair

path = '/content/gdrive/MyDrive/Colab_Notebooks/FYP/model/input/'
sentiment_model = flair.models.TextClassifier.load('en-sentiment')

csv_files = glob.glob(path+'news/*.{}'.format('csv'))
print(csv_files)
# df_concat_2020 = pd.read_csv(csv_files[2])
df_concat = pd.concat([pd.read_csv(f) for f in csv_files], ignore_index=True)
df_concat = df_concat.sort_values(by=['created_at'])
df_concat['combined_text'] = df_concat['description']+df_concat['content']
```

```python
probs = []
sentiments = []

for tweet in df_concat['combined_text']:
    sentence = flair.data.Sentence(str(tweet))
    sentiment_model.predict(sentence)
    probs.append(sentence.labels[0].score)
    sentiments.append(sentence.labels[0].value)

df_concat['probability'] = probs
df_concat['sentiment'] = sentiments
```

```python
df_concat.to_csv(path+'sentiment/news_sentiment_updated.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import os
import glob

path_2 = '/content/gdrive/MyDrive/Colab_Notebooks/FYP/model/input/sentiment/'

#list all csv files only
csv_files = glob.glob(path_2+'*.{}'.format('csv'))
print(csv_files)

raw_sentiment_df = pd.concat([pd.read_csv(f) for f in csv_files], ignore_index=True)
raw_sentiment_df['wsentiments'] = raw_sentiment_df['sentiment'].map({'POSITIVE':1, 'NEGATIVE':-1})*raw_sentiment_df['probability']
sentiment_df = raw_sentiment_df.groupby('created_at')['wsentiments'].mean()
sentiment_df.sort_index(inplace=True)
```

```python
sentiment_df.to_csv(path_2+'sentiment_data.csv')
```
