```python
import pandas as pd
import numpy as np
import os
import glob

path = '/content/gdrive/MyDrive/Colab_Notebooks/data/'

#list all csv files only
csv_files = glob.glob(path+'*.{}'.format('csv'))
print(csv_files)
df_concat_2020 = pd.read_csv(csv_files[2])
#df_concat = pd.concat([pd.read_csv(f) for f in csv_files ], ignore_index=True)
df_concat_2020
```

```python
df_concat_2020 = df_concat_2020.sort_values(by=['created_at'])
```

```python
df_concat_2020['combined_text'] = df_concat_2020['description']+df_concat_2020['content']
```

```python
import flair

sentiment_model = flair.models.TextClassifier.load('en-sentiment')
```

```python
probs = []
sentiments = []

for tweet in df_concat_2020['combined_text']:
    print(tweet)
    sentence = flair.data.Sentence(str(tweet))
    sentiment_model.predict(sentence)
    probs.append(sentence.labels[0].score)
    sentiments.append(sentence.labels[0].value)

df_concat_2020['probability'] = probs
df_concat_2020['sentiment'] = sentiments
```

```python
df_concat_2020.to_csv('news_sentiment_2020.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import os
import glob

path = '/content/gdrive/MyDrive/Colab_Notebooks/sentiment/'

#list all csv files only
csv_files = glob.glob(path+'*.{}'.format('csv'))
print(csv_files)

df_sentiment_concat = pd.concat([pd.read_csv(f) for f in csv_files ], ignore_index=True)
df_sentiment_concat
```

```python
df_sentiment_concat['wsentiments'] = df_sentiment_concat['sentiment'].map({'POSITIVE':1, 'NEGATIVE':-1})*df_sentiment_concat['probability']
sentiment_df = df_sentiment_concat.groupby('created_at')['wsentiments'].mean()
```

```python
sentiment_df.sort_index()
```

```python
sentiment_df.to_csv('sentiment_data.csv')
```
